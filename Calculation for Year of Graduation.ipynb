```python
# Custome Function to Get Values of Passing Year
def convert(row):
    Academic = row['Academic Year']
    current = row['Current year']
    year = row['Year']
    if Academic == 0.0 or current == 0:
        year = year + 4
    elif Academic == 1.0 or current == 1:
        year = year + 3
    elif Academic == 2.0 or current == 2:
        year = year + 2
    elif Academic == 3.0 or current == 3:
        year = year + 1
    elif Academic == 4.0 or current == 4:
        year == year
    return year
```
